```
ar = [1,3,2,6,1,2]
k = 3
```

```
n, k=map(int,input().split())
ar = [int(x) for x in input().strip().split()]
count = 0
for i in range(len(ar)):
  for j in range(len(ar)):
    if (j<i) & ((ar[i] + ar[j])%k == 0):
        count += 1
print(count)
```

Time Complexity = O(n^2)

```
arr = [1, 4, 4, 5, 5, 3]
```

```
input()
arr = [int(x) for x in input().strip().split()]
max = 1
sol = []
uni = list(set(arr))
for x in uni:
  if arr.count(x) > max:
    max = arr.count(x)
for x in uni:
  if arr.count(x) == max:
    sol.append(x)
print(min(sol)) 
```

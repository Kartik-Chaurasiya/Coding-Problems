```
ar = [1,3,2,6,1,2]
k = 3
```

```
n, k=map(int,input().split())
ar = [int(x) for x in input().strip().split()]
count = 0
for i in range(len(ar)):
  for j in range(len(ar)):
    if (j<i) & ((ar[i] + ar[j])%k == 0):
        count += 1
print(count)
```

Time Complexity = O(n^2)

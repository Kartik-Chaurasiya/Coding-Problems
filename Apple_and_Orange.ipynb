Complete the countApplesAndOranges function in the editor below. It should print the number of apples and oranges that land on Sam's house, each on a separate line.

countApplesAndOranges has the following parameter(s):

s: integer, starting point of Sam's house location.

t: integer, ending location of Sam's house location.
a: integer, location of the Apple tree.

b: integer, location of the Orange tree.

apples: integer array, distances at which each apple falls from the tree.

oranges: integer array, distances at which each orange falls from the tree.

```
import numpy as np
```

```
def get_in():
    arr = input().strip().split()
    for i in range(len(arr)):
        arr[i] = int(arr[i])
    return arr
def num(arr, s, t):
    count = 0
    for x in arr:
        if s <= x <= t:
            count += 1
    return count
    
s, t = [x for x in get_in()]
a, b = [x for x in get_in()]
m, n= [x for x in get_in()]
apple = []
orange = []
apple.append(get_in())
orange.append(get_in())
# apple = np.add(apple, a)
# orange = np.add(orange, b)
apple_new = [item for sublist in apple for item in sublist]
orange_new = [item for sublist in orange for item in sublist]
for x in range(len(apple_new)):
    apple_new[x] = apple_new[x] + a
for x in range(len(orange_new)):
    orange_new[x] = orange_new[x] + b   
for x in ([apple_new, orange_new]):
    print(num(x, s, t))
```

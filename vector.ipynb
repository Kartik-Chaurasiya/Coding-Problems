```
import numpy as np
import matplotlib.pyplot as plt
import random
```

quiver use for plotting vectors

```
plt.quiver(0,0,3,3)
plt.show()
```

```
plt.quiver(0,0,3,3, scale_units = 'xy', angles= 'xy', scale = 1)
plt.show()
```

```
plt.quiver(0,0,3,3, scale_units = 'xy', angles= 'xy', scale = 1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plt.quiver(0,0,3,3, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.quiver(0,0,-3,3, scale_units = 'xy', angles= 'xy', scale = 1, color = 'g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
for i in range (10):
  x = random.randint(1,10)
  y = random.randint(1,10)
  plt.quiver(0,0,x,y, scale_units = 'xy', angles= 'xy', scale = 1, color = 'g')
  plt.quiver(0,0,-x,y, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
